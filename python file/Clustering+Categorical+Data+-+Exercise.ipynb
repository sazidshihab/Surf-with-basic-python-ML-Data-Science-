## Import the relevant libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as mat
import seaborn as sns
sns.set()
from sklearn.cluster import KMeans
```

## Load the data

Load data from the csv file: <i> 'Categorical.csv'</i>.

```python
data = pd.read_csv('/Users/sazid/Downloads/Categorical.csv')
data
```

Remove the duplicate index column from the dataset. 

## Map the data

Use the <i>'continent'</i> category for this analysis.

```python
x = data.iloc[:,3:]
x['continent']= x['continent'].map({'North America':0,'Asia':1 ,'Africa':2,'Europe':3,'South America':4,'Oceania':5,'Antarctica':6,
                                    'Seven seas (open ocean)':7})
```

## Select the features

```python
x
```

## Clustering

Use 4 clusters initially.

```python
kmean = KMeans(8)
kmean.fit(x)
```

## Clustering results

```python
cluster = kmean.fit_predict(x)

data1=data
data1['cluster']= cluster
data1
```

## Plot the data

```python
mat.scatter(data1['Longitude'], data1['Latitude'], c=data1['cluster'], cmap='rainbow')
```
